```python
# Reference Exemplar — canonical FLAT notebook layout
#
# This single cell demonstrates the full kbu notebook workflow:
#   1. %run util.py  — bootstraps session, path constants, guarded imports
#   2. Trivial computation (offline, no BERDL dependency)
#   3. session.cache.save() — persist artifact to .kbcache/blobs/
#   4. session.cache.load() — round-trip verification
#
# The committed blob
# (.kbcache/blobs/265fc4c1...083d0.json) was produced by this cell.
# Re-running the cell overwrites the blob in-place (content-hash is
# deterministic for the same dict) and regenerates catalog.sqlite locally.

%run util.py

# Trivial offline computation — no BERDL / KBase / network required.
input_a = 6
input_b = 7
result = input_a + input_b

artifact = {
    "description": "Reference exemplar: trivial addition result",
    "input_a": input_a,
    "input_b": input_b,
    "result": result,
}

# Persist via session cache (writes to .kbcache/blobs/<sha256>.json).
entry = session.cache.save("reference_result", artifact)
print(f"Saved artifact: {entry.blob_path} ({entry.n_bytes} bytes)")

# Round-trip load: verifies the cache returns the same dict.
loaded = session.cache.load("reference_result")
assert loaded == artifact, f"Cache round-trip mismatch: {loaded!r}"
print(f"Round-trip OK: {loaded}")

# Path constants sanity check (flat layout: notebooks/ one level below project root).
assert NOTEBOOK_DIR == (PROJECT_ROOT / "notebooks").resolve(), (
    f"NOTEBOOK_DIR {NOTEBOOK_DIR} is not PROJECT_ROOT/notebooks"
)
print(f"PROJECT_ROOT : {PROJECT_ROOT}")
print(f"NOTEBOOK_DIR : {NOTEBOOK_DIR}")
print(f"DATA_DIR     : {DATA_DIR}")
print("All checks passed.")
```
